```python
import numpy as np
import pandas as pd
import seaborn as sns
import io
import plotly.graph_objects as go
import random
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/water_potability.csv')
df.interpolate(inplace=True)
df['Solids']=df['Solids']/100
```

```python
X = df.drop('Potability', axis=1)
y = df['Potability']
sns.set(font_scale=0.8)
plt.figure(figsize=(5, 5), dpi=500)


sns.heatmap(df.corr(), annot=True, linewidths=.5, annot_kws={"size": 6})
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
imputed_X_train = my_imputer.fit_transform(X_train)
imputed_X_test = my_imputer.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
model_dt.fit(imputed_X_train,y_train)
pred1 = model_dt.predict(imputed_X_test)
dtc = accuracy_score(y_test,pred1)
print(dtc)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(n_estimators=200,min_samples_leaf=0.16, random_state=0)
model_rf.fit(imputed_X_train, y_train)
predR = model_rf.predict(imputed_X_test)
predR
```
